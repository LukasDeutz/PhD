## Population density for current-based synapses

Here, we want to study the dynamics of neuronal networks of multiple populations using population density techniques. Our first objective is to generalize the method introduced in [1] to apply to current based-synapses instead of delta-synapses. This allows us to include different synaptic time constants for different connections types (e.g. excitatory and inhibitory).  

# Leaky integrate-and-fire neuron

For simplicity, we first study the leaky integrate-and-fire neuron which state is determined by membrane potential $V$. The dynamics of $V$ are determined by   

\begin{equation} \label{eq: V}
\frac{dV}{dt}=-\frac{V -V_{\text{R}}}{\tau_m} + I(t),
\end{equation}

where $\tau_{m}$ is the membrane time constant and $I(t)$ is current associated with the neuron. If the membrane potential reach the threshold $V_{\text{T}}$, the neurons emits a spike and the membrane potential is the reset potential $V_{\text{R}}$.

The current is invoked by incoming spikes from other neurons neurons within the network or external sources. For an exponential current-based synapse, $I(t)$ is given by

\begin{equation} \label{eq: I}
\frac{dI}{dt}=-\frac{I}{\tau}+hs(t),
\end{equation}

where $\tau$ the synaptic time constant, $h$ is the efficacy and $s(t)$ are spikes arriving the respective synapse 

\begin{equation} \label{eq: s}
s(t) =\sum_{n}\delta(t-t_{n}),
\end{equation}

at times $t_n$. It is convenient to perform the variable transformation 

\begin{equation}
v=\frac{V-V_{\text{R}}}{V_{\text{T}}-V_{\text{R}}},\quad j=\frac{I}{V_{\text{T}}-V_{\text{R}}}
\end{equation}

such that \ref{eq: V} and \ref{eq: I} becomes 

\begin{equation}
\frac{\partial v}{\partial t}=-\frac{v}{\tau_m}+j,\quad0\leq v\leq1,
\end{equation}

and 

\begin{equation}
\frac{dj}{dt}=-\frac{j}{\eta}+gs(t),
\end{equation}

where we absorbed the factor $V_{\text{T}}-V_{\text{R}}$ into the synaptic time constant and the synaptic efficacy such that

\begin{equation}
\eta=\tau(V_{\text{T}}-V_{\text{R}}),\quad g=\frac{h}{V_{\text{T}}-V_{\text{R}}}.
\end{equation}


The state of a neuron is determined by the $2$-dimensional state vector $\boldsymbol{x}(t)=(v(t),\,j(t))$. We are interested in the statistics of $\boldsymbol{x}(t)$ averaged over a population of interconnected neurons. To do this, we will consider the incoming spike arrivals \ref{eq: s} as a stochastic quantity. 
In general, s(t) has time dependent statistics which depend on the firing rate distribution of the postsynaptic neurons connecting to the neuron of interest. Assume that a neuron receives an average of $K$ randomly chosen inputs and that the average firing rate within the population at time $t$ is $r(t)$. The neurons may also receive external inputs with rate $r_{\text{ext}}(t)$. Hence, on average, a neuron feels an impulse rate 

\begin{equation} \label{eq: sigma}
\sigma(t) = Gr(t) + r_{\text{ext}}(t).
\end{equation}

The time evolution of the density $\rho(v,j,t)$ is determined by the conservation law 

\begin{equation} \label{eq: rho}
\frac{\partial\rho(v,j,t)}{\partial t} =-\frac{\partial}{\partial\boldsymbol{x}}\boldsymbol{J}(v, j, t),
\end{equation}

where $\boldsymbol{J}$ is the probability flux which consists of two contributions. First, the flux $\boldsymbol{J}_1$ due to the deterministic part of neuron dynamics 

\begin{equation}
\boldsymbol{J}_1 =\begin{pmatrix}-\frac{1}{\tau_{m}}v+j\\
-\frac{1}{\eta}j(t)
\end{pmatrix}\rho(v,j,t),
\end{equation}

Second, the flux $\boldsymbol{J}_2$ due to the jumps of the current variable $j$ invoked by incoming spikes. The first entry of $\boldsymbol{J}_2$ is zero because spikes only invoke jumps in $j$ and not in $v$. The second entry $\boldsymbol{J}_2$ is given by the integral expression       

\begin{equation}
J_{22} =-\sigma(t)\int_{j}^{j-g}dI'\,\rho(j,j',t)
\end{equation}

The integral boundaries are a consequence of the fact that at each spike arrival, $j$ makes a jump to $j'=j+g$. Note that $J_{22}$ is proportional to the total input rate \label{eq: sigma}, i.e. frequency on which jumps occure. The total flux $\boldsymbol{J}$ is the sum of both contribution 

\begin{equation} \label{eq: J}
\boldsymbol{J} = \boldsymbol{J}_1 + \boldsymbol{J}_2
\end{equation}

Inserting the $\boldsymbol{J}$ into the conservation law \ref{eq: rho} yields 

\begin{equation} \label{eq: rho 2}
\frac{\partial\rho}{\partial t}=\frac{1}{\tau_{m}}\frac{\partial(v\rho)}{\partial V}-j\frac{\partial\rho}{\partial v}+\frac{1}{\eta}\frac{\partial(j\rho)}{\partial j}+\sigma(t)\left[\rho(j-g)-\rho(j)\right]
\end{equation}

The average firing rate $r(t)$ at a given time $t$ must be equal to the flux passing through the threshold at $v=1$. Thus, we need to integrate over entire $j$ domain keeping $v$ fixed 

\begin{equation}
r(t)=\int_{-\infty}^{\infty}dj\,\boldsymbol{J}(1,j)\boldsymbol{e}_{1}.
\end{equation}

Inserting \ref{eq: J} yields 

\begin{equation} \label{eq: r}
r(t) = \int_{-\infty}^{\infty}dj\,\left(-\frac{1}{\tau_{m}}+j\right)\rho(1,j,t)
\end{equation}

The membrane potential can not exceed threshold, i.e. the probability density must be zero $\rho(v,j,t) = 0$ for $v > 1$. What about the boundary condition $\rho(1,j,t)$ = ? 

-  Mathematical argument: If $\rho(1,j,t) \neq 0$, then $\rho(v,j,t)$ makes a jump at $v=1$ and the derivative with respect to $v$ would infinite $\Rightarrow$ $\rho(1,j,t)=0$
-  Physical argument: For delta-synapses, voltage jumps move membrane potential to the threshold by discrete jumps. Only at $1-h$, the membrane potential will jump to the threshold. This event is highly unlikely, the chance that the membrane potential is very close to the threshold $1-\epsilon$, is more likely but still very unlikely. Thus, the probability density should go continuously to zero if $v$ approaches the threshold  

Substituting \ref{eq: rho} into \ref{eq: r} and solving for $r(t)$ yields

\begin{equation}
r(t) =\frac{r_{\text{ext}}(t)\int_{-\infty}^{\infty}dj\left(-\frac{1}{\tau_{m}}+j\right)\rho(1,j,t)}{1-G\int_{-\infty}^{\infty}dj\left(-\frac{1}{\tau_{m}}+j\right)\rho(1,j,t)}.
\end{equation}

Inserting into $\ref{rho 2}$, we arrive at 

\begin{equation}
\frac{\partial\rho}{\partial t}=\frac{1}{\tau_{m}}\frac{\partial(v\rho)}{\partial V}-j\frac{\partial\rho}{\partial v}+\frac{1}{\eta}\frac{\partial(j\rho)}{\partial j}+\frac{r_{\text{ext}}(t)}{1-G\int_{-\infty}^{\infty}dj\left(-\frac{1}{\tau_{m}}+j\right)\rho(1,j,t)}\left[\rho(j-g)-\rho(j)\right]
\end{equation}
